```python
import numpy as np
import math
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm


x = np.array([69, 74, 68, 73, 72, 87, 66, 70, 74, 68, 72, 89, 74, 67, 66, 71, 72, 75, 76, 78, 65])
y = np.array([163, 175, 165, 135, 172, 160, 115, 157, 200, 132, 140, 165, 285, 92, 142, 150, 165, 182, 212, 220, 134])

# Q- 2 (c)
plt.figure(figsize=(10,10))
loc = x.mean()
scale = x.std()
pdf = stats.norm.pdf(x, loc=loc, scale=scale)
plot = sns.lineplot(x=x, y=pdf)
plot.set_title("PDF Normal Distribution")
plot.set_xlabel("x")
plot.axvline(x.mean())
plot.set_ylabel("Probability Density")
```
